```python
import pandas as pd
DATA_DIR = "../data/"
TRAINDATA_DIR = "../traindata/"
```

```python
#read tweets

name_tweets = DATA_DIR + "clean_tweets/clean_tweets.json"
name_tweets_before = DATA_DIR + "clean_tweets/clean_tweets_before.json"
name_tweets_after = DATA_DIR + "clean_tweets/clean_tweets_after.json"

with open(name_tweets, "r") as file:
    tweets_df = pd.read_json(file, orient='index')
    
with open(name_tweets_before, "r") as file:
    before_tweets_df = pd.read_json(file, orient='index')
    
with open(name_tweets_after, "r") as file:
    after_tweets_df = pd.read_json(file, orient='index')
    
print(tweets_df.describe())
```

```python
#read stock data

name_stock = DATA_DIR + "DJI-20years-daily-change.json"

with open(name_stock, "r") as file:
    stock_df = pd.read_json(file, orient="split")
    
print(stock_df.describe())
```

```python
def store_time_stats(t, time_bins):
    if t.year not in time_bins:
        time_bins[t.year] = 1
    else: time_bins[t.year] += 1
       
    ym = str(t.year) + "-" + str(t.month)
    if ym not in time_bins:
        time_bins[ym] = 1
    else: time_bins[ym] += 1
```

```python
#assumes that the stocks are ordered by date

#DEPRECATED
def find_right_timestamp_index(start_ind, time):
    for i in range(start_ind, stock_df.shape[0]):
        date = stock_df.index[i]
        if time.date() != date.date():
            continue
        else: 
            #print ("right date!", date, time)
            return i
    return None
```

```python
#method to combine the data

def combine_data(df):
    classified = pd.DataFrame(columns=["timestamp", "text", "tweet", "value change"], index=range(df.shape[0]))
    print("classified shape:", classified.shape, "\ntweets_df shape:", df.shape)
    time_bins = {}
    latest_date_ind = 0

    for i, t in enumerate(df["timestamp"]):    
        #count how many tweet per year and month
        store_time_stats(t, time_bins)

        #fetch the change of the tweet's day
        change = stock_df.loc[t.date(), "daily change"]
    
        #save to datastruct
        classified["timestamp"][i] = t
        classified["text"][i] = df["text"][i]
        classified["tweet"][i] = df["tweet"][i]
        classified["value change"][i] = change
    
        if i%500==0: print("progress ", i/classified.shape[0])
            
    return classified, time_bins
```

```python
classified, time_bins = combine_data(tweets_df)
before_classified, before_time_bins = combine_data(before_tweets_df)
after_classified, after_time_bins = combine_data(after_tweets_df)
```

```python
# Suffling and splitting the dataframe into train and valid dataframes
from sklearn.model_selection import train_test_split
train, valid = train_test_split(classified, test_size=0.15)
before_train, before_valid = train_test_split(before_classified, test_size=0.15)
after_train, after_valid = train_test_split(after_classified, test_size=0.15)
```

```python
# Method to make a txt-file out of dataframe for FastText
def save_txt(dataframe, filename):
    learning_data = ""
    for i in range(0, len(dataframe)):
        ind = dataframe.index[i]
        change = dataframe["value change"][ind]
        text = dataframe["text"][ind]
        
        row = "__label__" + change + " " + text
        learning_data += row + "\n"
    print("thousand first characters\n", learning_data[:1000])
    with open(TRAINDATA_DIR + filename, "w") as file:
        file.write(learning_data)
```

```python
save_txt(train, "train.txt")
save_txt(valid, "valid.txt")
save_txt(before_train, "before_train.txt")
save_txt(before_valid, "before_valid.txt")
save_txt(after_train, "after_train.txt")
save_txt(after_valid, "after_valid.txt")
```
